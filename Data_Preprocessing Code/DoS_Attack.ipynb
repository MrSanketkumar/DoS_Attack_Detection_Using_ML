```python
missing_data= df.isnull()
for column in df :
  print(column)
  print(missing_data[column].value_counts())
  print(" ")
```

```python
df=df.dropna()
df=df.drop('Flow_Bytes_Sec',axis='columns')
df=df.drop('Flow_Packets_Sec',axis='columns')
columns=df.columns.tolist()
```

```python
df.shape
```

```python
df.columns
```

```python
df.corr()
```

```python

# other than ports column, normalize all featuers between zero and one
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
for i in columns:
  try:
    if (i== 'Destination_Port'):
      continue
    else:
      df[[f"{i}"]] = scaler.fit_transform(df[[f"{i}"]])
  except ValueError:
    continue
```

```python

# delete some columns which all of their values are zero

for k in range(len(columns)):
  if(df[columns[k]].min()==0 and df[columns[k]].max()==0):
    df=df.drop(columns[k],axis='columns')
```
